It is also called Transfer Function. The _activation function_ of a node defines the output of that node given an input or set of inputs. [wiki](https://en.wikipedia.org/wiki/Activation_function)


### sigmoid function, logistic or soft step

```python
%display latex
```

```python
f(x) = 1/(1+exp(-x))
df = f.derivative()
# df = f(x)(1-f(x))
df
```

```python
plot(f, -10, 10)
```

```python
plot(df, -10, 10)
```

### TanH

```python
f(x) = tanh(x)
df = f.derivative()
#df = 1 - f(x)^2
df
```

```python
plot(f, -10, 10)
```

```python
plot(df, -10, 10)
```

### ReLU

```python
f(x) = piecewise([((-1000,0), 0), ([0, 1000], x)]);  f
```

```python
plot(f, -10, 10)
```

### Leaky ReLU

```python
f(x) = piecewise([((-1000,0), 0.01*x), ([0, 1000], x)]);  f
```

```python
plot(f, -10, 10)
```

### Step Function

```python
step(x) = piecewise([((-1000,0), 0), ([0, 1000], 1)]);  step
```

```python
plot(step, -10, 10)
```
